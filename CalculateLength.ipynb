```python
import pandas as pd

df = pd.read_csv(r'C:\Users\jdelhoyo\PhD\Study cases\Genissiat\RV Characterization\repo-github\data\RV_For_RF4n1.csv')
df
```

```python
df_columns = {column.strip().lower().replace(' ', '_').replace('-', '_'): column for column in df.columns}

sub_basin_col = next((df_columns[key] for key in df_columns if 'sub_basin' in key or 'subbasin' in key), None)
length_col = next((df_columns[key] for key in df_columns if 'length' in key or 'lentgh' in key), None)

if sub_basin_col is None:
    raise KeyError(f"Could not find a Sub_Basin column. Available columns: {list(df.columns)}")
if length_col is None:
    raise KeyError(f"Could not find a Length column. Available columns: {list(df.columns)}")

length_by_sub_basin = (
    df.groupby(sub_basin_col, as_index=False)[length_col]
      .sum()
      .rename(columns={sub_basin_col: 'Sub_basin', length_col: 'Length_m'})
)

length_by_sub_basin['Length_m'] = length_by_sub_basin['Length_m'] / 2
length_by_sub_basin['Length_km'] = length_by_sub_basin['Length_m'] / 1000
length_by_sub_basin[['Sub_basin', 'Length_km']]
```
